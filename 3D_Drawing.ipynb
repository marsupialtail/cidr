```python
# Sketch of the 3D Plot


import numpy as np
import matplotlib.pyplot as plt

# Create the x and y values. We use np.linspace to avoid 0 as log(0) is undefined.
# x = np.linspace(0.1, 10, 100)
# y = np.linspace(0.1, 10, 100)

# # Use np.meshgrid to create a 2D grid of x and y values
# X, Y = np.meshgrid(x, y)
# Z = np.log(Y)

# x2 = np.linspace(10, 20, 100)
# y2 = np.linspace(0.1, 10, 100)
# X2, Y2 = np.meshgrid(x2, y2)
# Z2 = 2 * np.log(Y)


x = np.linspace(0.1, 10, 100)
y = np.linspace(0.1, 10, 100)

# Use np.meshgrid to create a 2D grid of x and y values
X, Y = np.meshgrid(x, y)
Z = 100 -np.log(Y)

x2 = np.linspace(10, 20, 100)
y2 = np.linspace(0.1, 10, 100)
X2, Y2 = np.meshgrid(x2, y2)
Z2 = 100- 2 * np.log(Y)


# Create the figure
fig = plt.figure(figsize=(15, 9))
ax = fig.add_subplot(111, projection='3d')

# Plot surfaces
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.plot_surface(X2, Y2, Z2, cmap='viridis')

# Set labels
ax.set_xlabel('RAM (GB)', labelpad=-10, size=16)
ax.set_ylabel('# of CPUs', labelpad=-10, size=16)
ax.set_zlabel('Runtime (s)', labelpad=-10, size=16)


# Set the viewing angle so that the X axis is pointing towards the viewer
ax.view_init(elev=15, azim=-15)

# # Set the axes to be bold
ax.xaxis.label.set_weight('bold')
ax.yaxis.label.set_weight('bold')
ax.zaxis.label.set_weight('bold')

#Remove the ticks from the axes
ax.xaxis.set_tick_params(width=0)
ax.yaxis.set_tick_params(width=0)
ax.zaxis.set_tick_params(width=0)
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])



# Show the plot
plt.savefig('3Dfigure.png', dpi=300)
plt.show()
```
